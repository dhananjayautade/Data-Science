```python
#spliting the data by comma of each row
data_split = list(data["products"].apply(lambda x:x.split(',')))
data_split
```

# Building Model
## Apriori

```python
# Importing the required library 
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(data_split).transform(data_split)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
#lets check the out result for our model
df1 = apriori(df,min_support=0.01,use_colnames=True)
df1
```
